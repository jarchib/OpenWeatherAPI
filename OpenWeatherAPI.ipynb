```python
##### Pulling data from API #####
import pandas as pd
from datetime import datetime 

datafinal = []
dataF = pd.DataFrame(datafinal)
loc = ['Anchorage, USA',
       'Buenos Aires, Argentina',
       'Sao Jose dos Campos, Brazil',
       'San Jose, Costa Rica',
       'Nanaimo, Canada',
       'Ningbo, China',
       'Giza, Egypt',
       'Mannheim, Germany',
       'Hyderabad, India',
       'Tehran, Iran',
       'Bishkek, Kyrgyzstan',
       'Riga, Latvia',
       'Quetta, Pakistan',
       'Warsaw, Poland',
       'Dhahran, Saudia Arabia',
       'Madrid, Spain',
       'Oldham, England']



for i in loc:
    import pprint 
    import requests 
    api_key = '454ae99306ab72083d5f8c0309168efc' 
    URL = 'https://api.openweathermap.org/data/2.5/forecast?' 
    URL = URL + f'q={i}&units=metric&appid=' + api_key 

    response = requests.get( URL )
    data = response.json() 
    
    temp_max = []
    temp_min = []
    datetime2 = []
    for j in range(0,40):
        temp_max.append(data ['list'][ j ]['main']['temp_max'])
        temp_min.append(data ['list'][ j ]['main']['temp_min'])
        datetime2.append(data ['list'][ j ]['dt_txt'])
        
    min_series = pd.Series(temp_min)
    max_series = pd.Series(temp_max)
    date_series = pd.Series(datetime2)
    frame = {'DateTime': date_series, 'MinTemp' : min_series, 'MaxTemp' : max_series }
    data1 = pd.DataFrame(frame)
    
    new=data1["DateTime"].str.split(" ", n = 1, expand = True)
    data1["Date"]=new[0]
    data1["Time"]=new[1]
    data1.drop(columns=["DateTime"], inplace = True)
    
    res=data1.groupby(['Date'], as_index=False)['MinTemp'].min()
    res2=data1.groupby(['Date'], as_index=False)['MaxTemp'].max()
    
    data2 = pd.concat([res, res2], axis=1, join='inner')
    
    if (data2.iloc[0,0] == datetime.today().strftime('%Y-%m-%d')):
        data3 = data2.drop(data2.index[0])
        data3.drop(columns=["Date"], inplace = True) 
        data4 = pd.DataFrame({'location': [i],
                      'min1': [data3['MinTemp'][1]],
                      'max1': [data3['MaxTemp'][1]],
                      'min2': [data3['MinTemp'][2]],
                      'max2': [data3['MaxTemp'][2]],
                      'min3': [data3['MinTemp'][3]],
                      'max3': [data3['MaxTemp'][3]],
                      'min4': [data3['MinTemp'][4]],
                      'max4': [data3['MaxTemp'][4]],
                      'min5': [data3['MinTemp'][5]],
                      'max5': [data3['MaxTemp'][5]],})
    else:
        data2.drop(columns=["Date"], inplace = True) 
        data4 = pd.DataFrame({'location': [i],
                      'min1': [data2['MinTemp'][0]],
                      'max1': [data2['MaxTemp'][0]],
                      'min2': [data2['MinTemp'][1]],
                      'max2': [data2['MaxTemp'][1]],
                      'min3': [data2['MinTemp'][2]],
                      'max3': [data2['MaxTemp'][2]],
                      'min4': [data2['MinTemp'][3]],
                      'max4': [data2['MaxTemp'][3]],
                      'min5': [data2['MinTemp'][4]],
                      'max5': [data2['MaxTemp'][4]],})
    
    data4['min avg'] = round(data4[['min1', 'min2', 'min3', 'min4', 'min5']].mean(axis=1),2)
    data4['max avg'] = round(data4[['max1', 'max2', 'max3', 'max4', 'max5']].mean(axis=1),2)
    data4
    
    dataF = dataF.append(data4, ignore_index=True)
    
dataF.to_csv(r'C:\Users\jessa\OneDrive\Documents\MASTERclasses\Python\OpenWeather.csv', index=False)
```
